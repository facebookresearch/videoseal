```python
import pandas as pd
import glob
from IPython.display import display, HTML
import base64

files = {
    "img": [i for i in csvs if "is_video=false" in i],
    "vid": [i for i in csvs if "is_video=true" in i]
}

# Initialize empty dictionaries to store the results
img_results = {}
vid_results = {}

# Load CSVs and calculate mean and std for images and videos
for file_type, files_list in files.items():
    results_dict = img_results if file_type == "img" else vid_results
    for file in files_list:
        try:
            df = pd.read_csv(file)
            file = os.path.dirname(file)
            if not df.empty:
                mean_std_df = df.describe().loc[["mean", "std"]]
                flipped_df = mean_std_df.transpose()
                results_dict[file] = flipped_df  # Store the result with the file name as key
            else:
                print(f"Warning: {file_type.capitalize()} file '{file}' is empty.")
        except pd.errors.EmptyDataError:
            print(f"Warning: {file_type.capitalize()} file '{file}' is empty.")
            
# Compare tables between all image models
all_img_dfs = []
for file, df in img_results.items():
    df['Mean+-Std'] = df.apply(lambda row: f"{row['mean']:.4f}+-{row['std']:.4f}", axis=1)
    dir_path = "/".join(glob.glob(file + "/*")[0].split("/")[:-1])
    img_wm_file = glob.glob(dir_path + "/*_wm.png")[0]
    img_diff_file = glob.glob(dir_path + "/*_diff.png")[0]
    with open(img_wm_file, "rb") as f:
        img_wm_data = f.read()
    img_wm_base64 = base64.b64encode(img_wm_data).decode("utf-8")
    with open(img_diff_file, "rb") as f:
        img_diff_data = f.read()
    img_diff_base64 = base64.b64encode(img_diff_data).decode("utf-8")
    df.loc['img_wm'] = f"<img src='data:image/png;base64,{img_wm_base64}' width='200' height='150'>"
    df.loc['img_diff'] = f"<img src='data:image/png;base64,{img_diff_base64}' width='200' height='150'>"
    all_img_dfs.append(df[['Mean+-Std']])
if all_img_dfs:
    all_img_comparison = pd.concat(all_img_dfs, axis=1)
    all_img_comparison.columns = list(img_results.keys())
# Compare tables between all video models
all_vid_dfs = []
for file, df in vid_results.items():
    df['Mean+-Std'] = df.apply(lambda row: f"{row['mean']:.4f}+-{row['std']:.4f}", axis=1)
    dir_path = "/".join(glob.glob(file + "/*")[0].split("/")[:-1])
    vid_wm_file = glob.glob(dir_path + "/*_wm.mp4")[0]
    vid_diff_file = glob.glob(dir_path + "/*_diff.mp4")[0]
    with open(vid_wm_file, "rb") as f:
        vid_wm_data = f.read()
    vid_wm_base64 = base64.b64encode(vid_wm_data).decode("utf-8")
    with open(vid_diff_file, "rb") as f:
        vid_diff_data = f.read()
    vid_diff_base64 = base64.b64encode(vid_diff_data).decode("utf-8")
    df.loc['vid_wm'] = f"<video width='200' height='150' controls><source src='data:video/mp4;base64,{vid_wm_base64}' type='video/mp4'>Your browser does not support the video tag.</video>"
    df.loc['vid_diff'] = f"<video width='200' height='150' controls><source src='data:video/mp4;base64,{vid_diff_base64}' type='video/mp4'>Your browser does not support the video tag.</video>"
    all_vid_dfs.append(df[['Mean+-Std']])
if all_vid_dfs:
    all_vid_comparison = pd.concat(all_vid_dfs, axis=1)
    all_vid_comparison.columns = list(vid_results.keys())
# Save HTML table to file
with open("image_results.html", "w") as f:
    f.write(all_img_comparison.to_html(escape=False))
with open("video_results.html", "w") as f:
    f.write(all_vid_comparison.to_html(escape=False))
```
